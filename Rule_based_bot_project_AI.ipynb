```python
import random #for dealing with random tasks
import re #for text patterns

class myBotfunctionalities:
    
  stop_conv = ("exit", "stop", "bye" , "goodbye", "quit")
  negative_resp = ("no", "not", "nope", "not a chance", "sorry")

  def __init__(self):
    self.query_responses = {
        'querying_about_internship_details' : r'.*\s*details.*',
        'querying_about_duration' : r'.*\s*duration.*',
        'about_stipend' : r'.*\s*stipend.*',
        'about_tasks' : r'.*\s*tasks.*',
        'Giving_ThankYou' : r'.*\s*ThankYou.*',
        'general_query' : r'.*\s*how.*'
    }

  def greetings(self):
    self.name = input("ChatBot Ally:  Hi!Welcome to this Rule Based Chatbot.I am Ally\nWhat's your name?\nYou  :")
    respond = input(f"ChatBot Ally:   Hi {self.name}, What would you like me to assist for?\nYou :  ")
    if respond in self.negative_resp:
        print("ChatBot Ally:  Oops!Seems Like you don't have any query\nYou  :")
        return 
    self.chat()

  def stop_conversation(self,resp):
    for command in self.stop_conv:
        if command in resp:
            print("ChatBot Ally:  Okay, Have a nice day! and thanks for reaching out.\nYou  :")
            return True
        return False
    
  def chat(self):
    response = input("ChatBot Ally:  Okay, How can I help You :) \nYou  :").lower()
    while not self.stop_conversation(response):
        response = input(self.identify_resp(response))

  def identify_resp(self,response):
    for intent, regex_pattern in self.query_responses.items():
        found_resp = re.search(regex_pattern,response)
        if found_resp and intent == 'querying_about_internship_details':
            return self.querying_about_internship_details()
        elif found_resp and intent == 'querying_about_duration':
            return self.querying_about_duration()
        elif found_resp and intent == 'about_stipend':
            return self.about_stipend()
        elif found_resp and intent == 'about_tasks':
            return self.about_tasks()
        elif found_resp and intent == 'general_query':
            return self.general_assis()
        elif found_resp and intent == 'Giving_ThankYou':
            return self.my_pleasure()
    return self.not_found_query()

  def querying_about_internship_details(self):
    reply = ("ChatBot Ally:  The Internship Will be based upon the domain which you've selected while applying on CodSoft Internships.\nYou  :",
             "ChatBot Ally:  All details will be shared with you on your provided email\nYou  :")
    return random.choice(reply)

  def querying_about_duration(self):
    reply = ("ChatBot Ally:  The duration for the Internship will be 1 month\nYou  :")
    return reply

  def about_stipend(self):
    reply = ("ChatBot Ally:  No, This is an Unpaid Internship in which we will provide you an offer letter and a certifiacte of completion\nYou  :",
             "ChatBot Ally:  This is Totally Unpaid Internship\nYou  :")
    return random.choice(reply)

  def about_tasks(self):
    reply = ("ChatBot Ally:  You're supposed to perform at least three tasks under this which CodSoft will provide you and you've to  submit them before the deadline\nYou  :",
             "ChatBot Ally:  You'will get five tasks and you've to perform at least three of them\nYou  :")
    return random.choice(reply)

  def general_assis(self):
    reply = ("ChatBot Ally:  Would You like me to assist you further?\nYou  :",
             "ChatBot Ally:  What else you would like to do?\nYou  :",
             "ChatBot Ally:  Do you have some more query?\nYou  :")
    return random.choice(reply)
  
  def my_pleasure(self):
     reply = ("ChatBot Ally:  My Pleasure\nYou  :")
     return reply

  def not_found_query(self):
    reply = ("ChatBot Ally:  Oh!Sorry I didn't get it\nYou  :",
             "ChatBot Ally:  Can You Please Rephrase it?\nYou  :",
             "ChatBot Ally:  Sorry, I'm not able to assist you at the moment\nYou  :")
    return random.choice(reply)


chatbot = myBotfunctionalities()
chatbot.greetings()
```
