```python
from sklearn.metrics import precision_score, recall_score, accuracy_score
```

```python
def loss(y_true, y_pred):
  pred = precision_score(y_true,y_pred)
  rec = recall_score(y_true,y_pred)
  acc = accuracy_score(y_true,y_pred)

  print(pred)
  print(rec)
  print(acc)
```
