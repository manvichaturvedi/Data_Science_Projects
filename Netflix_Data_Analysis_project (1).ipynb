**Data Cleaning**

**Handling Missing or null values Values**

```python
#creator column missing values handling.

df['creator'] = df['creator'].replace(np.nan, 'unknown')
```

```python
#genres column missing values handling.

df['genres'] = df['genres'].replace(np.nan, 'data unavailable')
```

```python
#country column missing values handling.

df['country'] = df['country'].replace(np.nan, 'unknown')
```

```python
df.isnull().sum()
```

**Inference from handling missing values**


*   Now , the datset does not contains any missing value.
*   The dataset is free from any duplicacy and null value.




Data Visualization
---



```python
plt.figure(figsize=(15,6))
plt.title("Percentage of Netflix Movie Types that are either Movies or TV Shows",fontsize=20,color='blue')
g = plt.pie(df.type.value_counts(),explode=(0.025,0.025), labels=df.type.value_counts().index, colors=['orange','pink'],autopct='%1.2f%%')
plt.show()
```

```python
df['creator'].value_counts().head(10)
```

```python
order=df.rating.value_counts()
order
```

```python
plt.figure(figsize=(10, 8))
sns.countplot(y='rating', data=df, order=order.index.to_list(), palette='Blues')
plt.title('Number of Content Titles by Rating', fontsize=21);

```

```python
country = df[-(df.country == 'unknown')]
plt.figure(figsize=(12,10))
plt.title('Top 10 Countries on the basis of Content Creation',fontsize=25)
sns.countplot(data=country,y='country',order=country['country'].value_counts().index[0:10],palette='Accent')
```

```python
plt.figure(figsize=(12,6))
plt.title('Top 10 Movie producing countries',fontsize='20')
Movie_countries = df[df['type']=='Movie']
sns.countplot(y='country', order = df['country'].value_counts().index[0:10],data=Movie_countries,color='yellow')

plt.figure(figsize=(12,6))
plt.title('Top 10 TV Shows producing countries',fontsize='20')
tvshow_countries = df[df['type']=='TV Show']
sns.countplot(y='country', order = df['country'].value_counts().index[0:10],data=tvshow_countries,color='purple')
```

```python
plt.figure(figsize=(20,10))
plt.xticks(rotation=90)
creator = df[-(df.creator == 'unknown')]
plt.title('Top 10 directors  producing content',fontsize='20')
sns.countplot(data=creator,x=df['creator'],order=df['creator'].value_counts().index[0:20],palette='mako')
plt.show()
```

```python
sns.countplot(data=df,y="type",hue="rating")
plt.show()
```

```python
labels = df['year'].value_counts().index
sizes = df['year'].value_counts()
plt.figure(figsize=(8,8))
plt.pie(sizes,labels=labels,rotatelabels=False,autopct='%1.1f%%')
plt.title('Year',color="Blue",fontsize=20)
plt.show()
```
